```python
p=df.iloc[:,2:3].values
```

# Training and Testing Data 

```python
from sklearn.model_selection import train_test_split
p_train,p_test,m_train,m_test = train_test_split(p,m,random_state=0)
```

```python
df.dropna(subset=["Tweet Polarity","Result"],inplace=True)
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

```python
m_train=np.nan_to_num(m_train)
p_train=np.nan_to_num(p_train)
```

```python
linreg=model.fit(p_train,m_train)
```

# Calculating the score value of test and train values

```python
print('R-squared score (training): {:.3f}'
     .format(model.score(p_train,m_train)))
print('R-squared score (test): {:.3f}'
     .format(model.score(p_test,m_test)))
```

```python
import matplotlib.pyplot as plt
```

# Linear Regression Model

```python
plt.figure(figsize=(7,8))
plt.scatter(p_train,m_train, marker= 'o', s=50, alpha=0.8)
plt.plot(p_train, linreg.coef_ * p_train + linreg.intercept_, 'r-')
plt.title('Linear Regression Model for Sentiment analysis of covid-19 tweets')
plt.xlabel('Polarity')
plt.ylabel('Result')
plt.legend(['Prediction','Result'])
plt.show()
```
